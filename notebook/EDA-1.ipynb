```python
# adding total and average
data['total'] = data['math_score'] + data['reading_score'] + data['writing_score']
```

```python
#average
data['average'] = data['total'] / 3
```

```python
data
```

# Taking out best performers

```python
full_maths = data[data['math_score']==100]['average'].count()
full_reading = data[data['reading_score']==100]['average'].count()
full_writing = data[data['writing_score']==100]['average'].count()
```

```python
print('Number of students who scored full marks in Mathematics: {}'.format(full_maths))
print('Number of students who scored full marks in reading: {}'.format(full_reading))
print('Number of students who scored full marks in writing: {}'.format(full_writing))
```

# Taking out worst performers

```python
less_maths = data[data['math_score'] < 15]['average'].count()
less_reading = data[data['reading_score'] < 15]['average'].count()
less_writing = data[data['writing_score'] < 15]['average'].count()

print('Number of students who scored less marks in Mathematics: {}'.format(less_maths))
print('Number of students who scored less marks in reading: {}'.format(less_reading))
print('Number of students who scored less marks in writing: {}'.format(less_writing))
```

# KDE Plots

```python
fig, axes = plt.subplots(1, 2, figsize = (20, 25))
plt.subplot(421)
sns.histplot(x = data['average'], bins=30, kde=True, color='g')
plt.subplot(422)
sns.histplot(x = data['average'], kde=True, hue=data['gender'])
plt.show()
```

```python
plt.subplots(1, 3, figsize = (20, 8))

# Both Genders
plt.subplot(141)
sns.histplot(x = data['average'], kde = True, hue=data['lunch'])

# Female
plt.subplot(142)
sns.histplot(data=data[data['gender']=='female'] , x = 'average', kde=True, hue='lunch')
plt.title('Female')

# Male
plt.subplot(143)
sns.histplot(data=data[data['gender']=='male'], x = 'average', kde=True, hue='lunch')
plt.title('Male')

plt.show()
```
